```python
class AudioDataset(Dataset):
    def __init__(self, root_dir, transform=None, sample_rate=16000):
        self.root_dir = root_dir
        self.transform = transform
        self.sample_rate = sample_rate
        self.audio_paths = []
        self.labels = []
        self.classes = sorted(os.listdir(root_dir))
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}

        for label in self.classes:
            class_dir = os.path.join(root_dir, label)
            for file in os.listdir(class_dir):
                if file.endswith((".wav", ".mp3")):
                    self.audio_paths.append(os.path.join(class_dir, file))
                    self.labels.append(self.class_to_idx[label])

    def __len__(self):
        return len(self.audio_paths)

    def __getitem__(self, idx):
        audio_path = self.audio_paths[idx]
        waveform, sample_rate = torchaudio.load(audio_path)
        
        # Resample if needed
        if sample_rate != self.sample_rate:
            resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=self.sample_rate)
            waveform = resampler(waveform)

        if self.transform:
            waveform = self.transform(waveform)

        label = self.labels[idx]
        return waveform, label

def get_class_distribution(dataset):
    class_counts = {}
    for _, label in dataset:
        if label in class_counts:
            class_counts[label] += 1
        else:
            class_counts[label] = 1
    return class_counts


def load_dataset(dataset_path):

    dataset = AudioDataset(root_dir=dataset_path, transform=None)
    print(f"Loaded {len(dataset)} samples from {dataset_path}")
    return dataset

# Replace this with your actual dataset value for 'd'

dataset = load_dataset("/media/data/hamid/raw_vowels/5FS")  # replace with actual path
class_names = dataset.classes
class_d=get_class_distribution(dataset)

print("Class distribution:", class_d)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

Classes = ['healthy', 'mild', 'severe']

for d in ["1AS", "1FS", "5AS", "5FS"]:
    print(f"\n=== Dataset {d} ===")

    CM_paths = [
        f"./cm_rese/confusion_matrix_ReSE_{d}_fold_1.csv",
        f"./cm_rese/confusion_matrix_ReSE_{d}_fold_2.csv",
        f"./cm_rese/confusion_matrix_ReSE_{d}_fold_3.csv"
    ]

    precisions, recalls, f1s, accuracies = [], [], [], []
    class_accuracies = {i: [] for i in range(len(Classes))}

    for path in CM_paths:
        cm = pd.read_csv(path, index_col=0).values

        y_true, y_pred = [], []
        for true_label in range(cm.shape[0]):
            for pred_label in range(cm.shape[1]):
                count = cm[true_label][pred_label]
                y_true.extend([true_label] * count)
                y_pred.extend([pred_label] * count)

        precision, recall, f1, _ = precision_recall_fscore_support(
            y_true, y_pred, labels=[0, 1, 2], zero_division=0
        )
        acc = accuracy_score(y_true, y_pred)

        precisions.append(precision)
        recalls.append(recall)
        f1s.append(f1)
        accuracies.append(acc)

        # Per-class accuracy: correct / total of each class
        for i in range(len(Classes)):
            class_total = np.sum(cm[i, :])
            class_correct = cm[i][i]
            class_acc = class_correct / class_total if class_total > 0 else 0.0
            class_accuracies[i].append(class_acc)

    precisions = np.array(precisions)
    recalls = np.array(recalls)
    f1s = np.array(f1s)
    accuracies = np.array(accuracies)


    for i, class_name in enumerate(Classes):
        print(f"\nClass: {class_name}")
        print(f"  Precision: {precisions[:, i].mean():.4f} Â± {precisions[:, i].std():.4f}")
        print(f"  Recall:    {recalls[:, i].mean():.4f} Â± {recalls[:, i].std():.4f}")
        print(f"  F1 Score:  {f1s[:, i].mean():.4f} Â± {f1s[:, i].std():.4f}")
        print(f"  Accuracy:  {np.mean(class_accuracies[i]):.4f} Â± {np.std(class_accuracies[i]):.4f}")

    # Save the metrics to a CSV file
    results = []
    for i, class_name in enumerate(Classes):
        results.append({
            "Class": class_name,
            "Precision": f"{precisions[:, i].mean():.4f} Â± {precisions[:, i].std():.4f}",
            "Recall": f"{recalls[:, i].mean():.4f} Â± {recalls[:, i].std():.4f}",
            "F1 Score": f"{f1s[:, i].mean():.4f} Â± {f1s[:, i].std():.4f}",
            "Accuracy": f"{np.mean(class_accuracies[i]):.4f} Â± {np.std(class_accuracies[i]):.4f}"
        })

    results_df = pd.DataFrame(results)
    results_df.to_csv(f"./results/metrics_ReSE_{d}.csv", index=False)
    print(f"Metrics saved to ./results/metrics_ReSE_{d}.csv")
```

```python
import parselmouth
import statistics
import numpy as np
from parselmouth.praat import call
import os
import pandas as pd
import librosa

def Acoustic_features(voiceID, f0min=75, f0max=500, unit="Hertz"):
    sound = parselmouth.Sound(voiceID)
    pitch = call(sound, "To Pitch", 0.0, f0min, f0max)
    meanF0 = call(pitch, "Get mean", 0 ,0, unit)
    harmonicity = call(sound, "To Harmonicity (cc)", 0.01, f0min, 0.1, 1.0)
    hnr = call(harmonicity, "Get mean", 0, 0)
    pointProcess = call(sound, "To PointProcess (periodic, cc)", f0min, f0max)

    localJitter = call(pointProcess, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
    localabsoluteJitter = call(pointProcess, "Get jitter (local, absolute)", 0, 0, 0.0001, 0.02, 1.3)
    rapJitter = call(pointProcess, "Get jitter (rap)", 0, 0, 0.0001, 0.02, 1.3)
    ddpJitter = call(pointProcess, "Get jitter (ddp)", 0, 0, 0.0001, 0.02, 1.3)

    localShimmer =  call([sound, pointProcess], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    localdbShimmer = call([sound, pointProcess], "Get shimmer (local_dB)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    apq3Shimmer = call([sound, pointProcess], "Get shimmer (apq3)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    aqpq5Shimmer = call([sound, pointProcess], "Get shimmer (apq5)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    
    return localabsoluteJitter, localJitter, rapJitter, ddpJitter, localdbShimmer, localShimmer, apq3Shimmer, aqpq5Shimmer, hnr, meanF0


def extract_mfcc(audio_path, sr=16000, n_mfcc=13):
    y, _ = librosa.load(audio_path, sr=sr)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfccs, axis=1)

folders = ["/media/data/hamid/with_pad/1S/1AS","/media/data/hamid/with_pad/2S/2AS","/media/data/hamid/with_pad/3S/3AS","/media/data/hamid/with_pad/4S/4AS","/media/data/hamid/with_pad/5S/5AS","/media/data/hamid/with_pad/6S/6AS","/media/data/hamid/with_pad/7S/7AS"]

# Update columns for individual formants
columns = [
    "path", "localabsoluteJitter", "localJitter", "rapJitter", "ddpJitter",
    "localdbShimmer", "localShimmer", "apq3Shimmer", "aqpq5Shimmer",
    "hnr", "fundamental_frequency"
] + [f"MFCC{i}" for i in range(13)] + ["audio_id", "segment_number"] # Add segment_number column

for folder in folders:
    print(f"Processing: {folder}")
    
    # Use a list to collect feature dictionaries, then create DataFrame once
    all_features_list = []

    for subfolder in os.listdir(folder):
        subfolder_path = os.path.join(folder, subfolder)
        for filename in os.listdir(subfolder_path):
            if filename.endswith(".wav"):
                try:
                    path = os.path.join(subfolder_path, filename)
                    features = {}

                    jitter_vals = Acoustic_features(path)
                    
                    mfccs = extract_mfcc(path)
                    
                    # Extract segment number for sorting
                    # Assuming filename is like "ID_segX.wav"
                    try:
                        segment_number = int(filename.split('_seg')[-1].split('.')[0])
                    except ValueError:
                        print(f"Warning: Could not parse segment number from {filename}. Setting to -1.")
                        segment_number = -1 # Or handle as per your needs

                    features.update({
                        "path": path,
                        "localabsoluteJitter": jitter_vals[0],
                        "localJitter": jitter_vals[1],
                        "rapJitter": jitter_vals[2],
                        "ddpJitter": jitter_vals[3],
                        "localdbShimmer": jitter_vals[4],
                        "localShimmer": jitter_vals[5],
                        "apq3Shimmer": jitter_vals[6],
                        "aqpq5Shimmer": jitter_vals[7],
                        "hnr": jitter_vals[8],
                        "pitch": jitter_vals[9],
                        "fundamental_frequency": jitter_vals[9],
                        "audio_id": filename,
                        "segment_number": segment_number
                    })

                    for i in range(len(mfccs)):
                        features[f"MFCC{i}"] = mfccs[i]

                    all_features_list.append(features)

                except Exception as e:
                    print(f"[Error Processing {filename}]: {e}")

    df = pd.DataFrame(all_features_list, columns=columns)

    # --- NaN Handling with bfill ---
    # 1. Sort the DataFrame by segment_number to ensure correct sequential fill
    df.sort_values(by=['audio_id', 'segment_number'], inplace=True) # Sort by audio_id first to group segments from the same source audio, then by segment_number

    # 2. Identify columns to apply bfill (all numeric columns except 'segment_number')
    # Exclude 'path' and 'audio_id' and 'segment_number'
    numeric_cols_to_fill = [col for col in df.columns if df[col].dtype in ['float64', 'int64'] and col not in ["segment_number"]]

    # Apply bfill. Using .groupby() + .bfill() is safer if you have multiple original audio files
    # within the same 'folder' and want to fill only within segments of the same original audio.
    # Assuming 'audio_id' contains a common prefix for segments of the same original audio (e.g., 'ID_segX')
    # Let's get the base ID (e.g., 'ID' from 'ID_seg0') for grouping
    df['base_audio_id'] = df['audio_id'].apply(lambda x: x.split('_seg')[0] if '_seg' in x else x)

    # Apply bfill within each group of base_audio_id
    for col in numeric_cols_to_fill:
        if df[col].isnull().any():
            df[col] = df.groupby('base_audio_id')[col].bfill()
            print(f"Applied bfill to '{col}' for folder {os.path.basename(folder)}")

    # After bfill, check for any remaining NaNs (e.g., if the *last* segment of a sequence had NaN)
    # If bfill leaves NaNs (which it will if NaNs are at the end of a group/series),
    # you might consider ffill or median imputation as a fallback.
    for col in numeric_cols_to_fill:
        if df[col].isnull().any():
            median_val = df[col].median()
            df[col] = df[col].fillna(median_val)
            print(f"Filled remaining NaNs in '{col}' with median for folder {os.path.basename(folder)}")

    # Drop the temporary 'base_audio_id' and 'segment_number' columns if you don't need them in the final CSV
    df.drop(columns=['base_audio_id', 'segment_number'], inplace=True)

    # --- End NaN Handling ---

    save_path = f"/media/data/hamid/AcousticFeatures_with_pad_2/{os.path.basename(folder)}.csv"
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    if os.path.exists(save_path):
        os.remove(save_path)
    df.to_csv(save_path, index=False)
    print(f"Saved features to {save_path}")

    # Optional: Print NaN counts after filling for verification
    nan_counts_after = df.isnull().sum()
    print(f"\nNaN counts for {os.path.basename(folder)} after filling:")
    print(nan_counts_after[nan_counts_after > 0]) # Print only columns with NaNs
```

```python
import parselmouth
import statistics
import numpy as np
from parselmouth.praat import call
import os
import pandas as pd
import librosa

def Acoustic_features(voiceID, f0min=75, f0max=500, unit="Hertz"):
    sound = parselmouth.Sound(voiceID)
    pitch = call(sound, "To Pitch", 0.0, f0min, f0max)
    meanF0 = call(pitch, "Get mean", 0 ,0, unit)
    harmonicity = call(sound, "To Harmonicity (cc)", 0.01, f0min, 0.1, 1.0)
    hnr = call(harmonicity, "Get mean", 0, 0)
    pointProcess = call(sound, "To PointProcess (periodic, cc)", f0min, f0max)

    localJitter = call(pointProcess, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
    localabsoluteJitter = call(pointProcess, "Get jitter (local, absolute)", 0, 0, 0.0001, 0.02, 1.3)
    rapJitter = call(pointProcess, "Get jitter (rap)", 0, 0, 0.0001, 0.02, 1.3)
    ddpJitter = call(pointProcess, "Get jitter (ddp)", 0, 0, 0.0001, 0.02, 1.3)

    localShimmer =  call([sound, pointProcess], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    localdbShimmer = call([sound, pointProcess], "Get shimmer (local_dB)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    apq3Shimmer = call([sound, pointProcess], "Get shimmer (apq3)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    aqpq5Shimmer = call([sound, pointProcess], "Get shimmer (apq5)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    
    return localabsoluteJitter, localJitter, rapJitter, ddpJitter, localdbShimmer, localShimmer, apq3Shimmer, aqpq5Shimmer, hnr, meanF0

def average_formant_frequency(voice_id, f0min=75, f0max=500):
    try:
        sound = parselmouth.Sound(voice_id)
        pointProcess = call(sound, "To PointProcess (periodic, cc)", f0min, f0max)
        formants = call(sound, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
        numPoints = call(pointProcess, "Get number of points")

        f1_list, f2_list, f3_list, f4_list = [], [], [], []

        for point in range(1, numPoints + 1):
            t = call(pointProcess, "Get time from index", point)
            f1 = call(formants, "Get value at time", 1, t, 'Hertz', 'Linear')
            f2 = call(formants, "Get value at time", 2, t, 'Hertz', 'Linear')
            f3 = call(formants, "Get value at time", 3, t, 'Hertz', 'Linear')
            f4 = call(formants, "Get value at time", 4, t, 'Hertz', 'Linear')

            if not np.isnan(f1): f1_list.append(f1)
            if not np.isnan(f2): f2_list.append(f2)
            if not np.isnan(f3): f3_list.append(f3)
            if not np.isnan(f4): f4_list.append(f4)

        if not all([f1_list, f2_list, f3_list, f4_list]):
            return None

        return statistics.median(f1_list + f2_list + f3_list + f4_list) / 4
    except Exception as e:
        print(f"[Formant Error] {voice_id}: {e}")
        return None

def extract_mfcc(audio_path, sr=16000, n_mfcc=13):
    y, _ = librosa.load(audio_path, sr=sr)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfccs, axis=1)

folders = ["/media/data/hamid/with_pad/1S/1AS","/media/data/hamid/with_pad/2S/2AS","/media/data/hamid/with_pad/3S/3AS","/media/data/hamid/with_pad/4S/4AS","/media/data/hamid/with_pad/5S/5AS","/media/data/hamid/with_pad/6S/6AS","/media/data/hamid/with_pad/7S/7AS"]

for folder in folders:
    print(f"Processing: {folder}")
    columns = [
    "path", "localabsoluteJitter", "localJitter", "rapJitter", "ddpJitter",
    "localdbShimmer", "localShimmer", "apq3Shimmer", "aqpq5Shimmer",
    "hnr", "pitch", "FundamentalFrequency"] + [f"MFCC{i}" for i in range(13)] + ["audio_id"]

    df = pd.DataFrame(columns=columns)

    for subfolder in os.listdir(folder):
        subfolder_path = os.path.join(folder, subfolder)
        for filename in os.listdir(subfolder_path):
            if filename.endswith(".wav"):
                try:
                    path = os.path.join(subfolder_path, filename)
                    features = {}

                    jitter_vals = Acoustic_features(path)
                    formant = average_formant_frequency(path)
                    mfccs = extract_mfcc(path)
                    


                    features.update({
                        "path": path,
                        "localabsoluteJitter": jitter_vals[0],
                        "localJitter": jitter_vals[1],
                        "rapJitter": jitter_vals[2],
                        "ddpJitter": jitter_vals[3],
                        "localdbShimmer": jitter_vals[4],
                        "localShimmer": jitter_vals[5],
                        "apq3Shimmer": jitter_vals[6],
                        "aqpq5Shimmer": jitter_vals[7],
                        "hnr": jitter_vals[8],
                        "pitch": jitter_vals[9],
                        "FundamentalFrequency": formant,
                        "audio_id": filename
                    })

                    for i in range(len(mfccs)):
                        features[f"MFCC{i}"] = mfccs[i]

                    features["audio_id"] = filename 

                    df.loc[len(df)] = [features[col] for col in columns]  # ensures correct order


                    if len(df) < 5:
                        print(df.tail(1))

                except Exception as e:
                    print(f"[Error Processing {filename}]: {e}")

    save_path = f"/media/data/hamid/AcousticFeatures_with_pad/{os.path.basename(folder)}.csv"
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    df.to_csv(save_path, index=False)
    print(f"Saved features to {save_path}")
```
